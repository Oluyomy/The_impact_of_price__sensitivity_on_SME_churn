```python
#scattered plots to show correlation between cons_12m and cons_gas_12m
sns.scatterplot(data=client_df, 
                x="cons_12m", 
                y="cons_gas_12m",
                alpha = 0.7)
plt.suptitle("Scatter Plot cons_12m to cons_gas_12m")
plt.show()
```

```python
#to remove non numerc column
numeric_client_df = client_df.select_dtypes(include=[np.number])
corr = numeric_client_df.corr()
corr
```

```python
#Heatmap showing correlation in different variables.
# Create a heat map with correlation data
sns.heatmap(data= corr, annot=True)
plt.title("Correlation Matrix Heatmap")
```

```python
cons_12m = client_df['cons_12m']
forecast_cons_12m = client_df['forecast_cons_12m']
forecast_price_energy_off_peak = client_df['forecast_price_energy_off_peak']
forecast_price_energy_peak = client_df['forecast_price_energy_peak']
```

```python
# Calculate price sensitivity for the entire DataFrame
change_in_consumption = (forecast_cons_12m - cons_12m) / cons_12m
change_in_price = (forecast_price_energy_peak - forecast_price_energy_off_peak) / forecast_price_energy_off_peak
client_df['price_sensitivity'] = abs(change_in_consumption / change_in_price)
```
